```python
# Question-2- Three students choose elective subjects.
# Find the common subjects chosen by all students using set operation.
# Concept Used: Set intersection(&)

student1={"python","physics","english","hindi"}
student2={"python","english","chemistry","feee"}
student3={"python","mathematics","english","beee"}
common_subjects=student1 & student2 & student3
print(common_subjects)
```

```python
#Question-3: One set contains students in the cricket club and another in the football club.
#Find students who are 1. In both clubs 2. in only one club
# Concept Used: Intersection, Symmetric difference
Cricket= {"Amit","Om","Sneha"}
Football= {"Priya", "Om","Aditya"}
print("Both clubs:", Cricket & Football)
print("Only one club:",Cricket^Football)
```

```python
# Question-4 Absent Students
# Given a set of all students and a set of present students, find the absent students.
Total={"Maahi","Sara","Rupsa","Priyal","Sijisha"}
Present={"Maahi","Sara","Priyal"}
print("Absent:",Total-Present)
```

```python
# Question-5 Remove Invalid Entries.
# A set contains valid courses codes. Remove a discotinued course code from the set.
# Concept Used: remove()/discard()
Courses={"Python","Calculus","Tinker_lab"}
Courses.discard("Calculus")
print("Updated Courses:",Courses)
```
